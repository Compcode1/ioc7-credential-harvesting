**IOC 7: Email Link to Fake HR Portal – Credential Harvesting Attempt**
Triage Category: Social Engineering / Obfuscated Link

**(1) Attacker Analogy**
Imagine receiving an email from your company’s HR department asking you to review your upcoming benefits package. The email is perfectly formatted, uses your name, and includes the correct logo. You click the link—it opens a professional-looking page with your company’s branding. But what you do not realize is that the entire site is fake, built by someone outside the company. The goal isn’t to inform you about your benefits—it’s to steal your login credentials when you enter them.

Softened Interpretation
This isn’t a brute-force attack or a malware drop. It’s a carefully constructed scam that uses trust and appearance to trick users into giving away their secrets. No antivirus alert, no strange process behavior—just a lie wrapped in familiarity.

**(2) Source of IOC (Telemetry Origin)**
The IOC originated from a Secure Web Gateway (SWG) log that flagged access to an uncategorized domain by a corporate user. Further inspection revealed the domain was newly registered, mimicked internal branding, and hosted a login form that redirected input to an external server.

Softened Interpretation
The first warning didn’t come from the user—it came from a filter in the system that noticed someone accessing a strange website. The site had no known category and was brand new, which raised an automated red flag.

**(3) Triage Framework Declaration**
Framework: Social Engineering / Obfuscated Link Triage
Canonical Toolkit Used:

Artifact Capture – Email and headers saved from user report.

URL Inspection – Performed via curl and browser sandbox.

User Investigation – Interviewed target user to clarify actions.

Isolation and Blocking – Domain blocked at SWG and DNS layers.

Softened Interpretation
We used our standard toolkit for phishing-style attacks. First, we captured the actual email that tricked the user. Then we tested the link in a safe environment. We also asked the user what they did, and finally, we blocked the domain from future access.

**(4) OS Layer with Key Behaviors**
Not Applicable (Not a host-based IOC)

**(5) Cross-Layer Interaction Pivots**
Not Applicable (No internal process execution or host linkage yet)

**(6) OSI Layer Relevance**
Layer 7 – Application Layer.
The entire attack occurred at the top of the OSI stack. DNS resolution, HTTP connection, and credential submission all occurred within the context of user-initiated browser traffic. The attacker exploited application layer behavior by crafting a convincing front-end and capturing user keystrokes via HTTPS POST to their own server.

Softened Interpretation
This wasn’t a deep-system trick—it all happened in the browser. The attacker didn’t dig into the machine; they just waited for the user to hand over information willingly.

**(7) Attacker Behavior Interpretation**
The attacker conducted a classic phishing attack using a highly tailored fake login page. The email was well-crafted, referencing HR terms, company-specific branding, and included no attachment—just a link. The link led to a site with a domain designed to look internal, e.g. corpbenefits-update[.]com, and hosted a login form matching the real portal.

The attacker didn’t use malware or exploits—only psychology. Once the victim entered credentials, the form sent the data to a remote endpoint for harvesting. No privilege escalation, no persistence, just one clean theft attempt.

Softened Interpretation
This attacker wasn’t trying to break in through the back door—they just knocked politely and waited for someone to open the door themselves. It’s low-tech but high-efficiency.

**(8) Defender Action Summary**
Captured and preserved full email and link artifacts.

Inspected the link and domain via sandboxed curl and browser analysis.

Confirmed credential harvesting mechanism via POST behavior.

Notified targeted users and reset passwords.

Blocked domain at SWG and internal DNS resolvers.

Created indicator-based alerting for similar future domains.

Softened Interpretation
We saved the email, tested the link, and confirmed it was a trap. Once we knew credentials were being stolen, we shut down access, reset passwords, and told everyone what happened to make sure it didn’t work again next time.

**(9) Attacker Strategy Notes**
This attacker favored a single-stage strategy: craft a believable lure, create a fake interface, and harvest credentials without triggering technical defenses. No payloads were dropped, no executables launched, and no EDR alerts raised.

Their strategy depended entirely on speed and deception. They aimed to collect credentials in a narrow window and disappear. Had the SWG not flagged the domain, the attack might have succeeded unnoticed. This underscores the importance of human trust and early-stage URL monitoring as frontline defense in credential theft.

**(10) Final Analyst Reflection and Softened Narrative**
This case is deceptively simple—no malware, no exploit code. And yet, its danger lies in how close it came to working. Phishing attacks like this bypass technical detection by staying within normal user behavior channels. This reinforces that some of the most dangerous attacks don’t look like attacks at all.

Softened Reflection:
Sometimes the scariest threats wear friendly faces. This attacker didn’t force entry—they asked politely, smiled, and waited for a click. That’s all it takes. This case reminds us that the most human parts of the system—curiosity, trust, routine—are also the most exploitable.

**(11) Who’s Who – Object Role Clarification**
Narrative Summary of Roles:

Attacker Email – The original phishing email disguised as an HR communication. No payload attached—just a link.

Phishing URL – The link embedded in the email pointing to the fake login page. Hosted on corpbenefits-update[.]com.

Target User – The employee who clicked the link and attempted to log in.

Secure Web Gateway (SWG) – The system that detected and logged the suspicious domain request.

DNS Resolver and Firewall – Internal network components used to confirm no other machines accessed the domain.

Conclusion / README
Case Title: IOC 7 – HR Email Phish with Fake Login Portal
Primary Threat: Credential harvesting via impersonated HR communication
Triage Type: Social Engineering / Obfuscated Link
Detection Vector: Secure Web Gateway domain categorization anomaly
Key Response Actions: Domain analysis, credential reset, domain block
Lessons Learned:

Even well-written phishing emails can bypass standard defenses.

Users must be trained to spot subtle signs of fakes.

Monitoring newly registered domains and uncategorized traffic is essential.

The attack surface includes employee trust—not just endpoints and ports.

This project exemplifies how quiet attacks on human behavior can be just as dangerous as malware-based intrusions. A solid social engineering triage protocol—and fast response—prevented credential loss in this case.
